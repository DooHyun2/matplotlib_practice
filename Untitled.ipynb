```python
import pandas as pd
import numpy as np
df = pd.read_csv("synthetic_LLZO.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["dop"], df["sigma"])
plt.xlabel("dopant fraction")
plt.ylabel("conductivity (sigma)")
plt.title("dop vs sigma")
plt.show()
```

```python
plt.scatter(df["density"], df["sigma"])
plt.xlabel("density")
plt.ylabel("conductivity")
plt.title("density vs sigma")
plt.show()
```

```python
df["invT"] = 1 / df["temp_k"]
df["log_sigma"] = np.log(df["sigma"])

plt.scatter(df["invT"], df["log_sigma"])
plt.xlabel("1/t (1/k)")
plt.ylabel("log(sigma)")
plt.title("Arrhenius plot")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 

x = df[["dop", "density", "gb_frac", "temp_c"]]
y = df["log_sigma"]

model = LinearRegression()
model.fit(x, y)

model.coef_, model.intercept_
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

X = df[["dop", "density", "gb_frac", "temp_c"]]
y = df["log_sigma"]

rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)
rf.fit(X, y)

pred = rf.predict(X)

print("R2:", r2_score(y, pred))
print("Feature importance:", rf.feature_importances_)
```

```python
import optuna

def objective(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 500)
    max_depth    = trial.suggest_int("max_depth", 3, 20)

    m = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=42
    )
    m.fit(X, y)
    return r2_score(y, m.predict(X))

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)

study.best_params, study.best_value
```
